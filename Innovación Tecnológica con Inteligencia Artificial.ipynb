# Sesión 4: Recopilación, Manejo y Uso de Datos

## Tenemos una base de datos simple: en una hoja de cálculo, tenemos varias tabs, una por evento. Cada Tab, tiene el 

* ## nombre de sesión
* ## nombre de asistentes
* ## appellido de asistentes
* ## correo de asistentes
* ## número de clics para entrar a la sesión

## Con el siguiente código, cargamos la base datos y con la información transformada para salvaguardar el anonimato de las personas y de este cliente. 

## Te muestro unos cuantos renglones de cada tab modificado para que veas la información.

```python
import numpy as np
import pandas as pd

from tqdm.notebook import tqdm

import numpy as np
import pandas as pd

from tqdm.notebook import tqdm

db = pd.read_excel('database.xlsx', sheet_name = None)
admin_tabs = ['Total Registrants']
event_tabs = list(set(db.keys()) - set(admin_tabs))

emails = db['Total Registrants']['Email'].unique()

for event in event_tabs:
    display(event, db[event].head(5))
```

## Con esta información, podemos analizar las tendencias de asistencia en los diferentes eventos; esencialmente, responder a la pregunta: si alguien se conectó al evento uno, ¿qué tan probable habrá sido conectarte al evento dos? ¿y al evento tres?, ¿y al cuatro?, cinco, seis, ...

## Con esto, podemos evaluar patrones de interés y darnos una idea de cómo es que los asistentes interactuaron entre sí. 

## El siguiente código va a generar una matriz de asistencia  y después, vamos a generar una tabla con estadística descriptiva para ver cómo se comportan los datos en cada evento de una manera resumida. 

```python
attendee_event_clicks_matrix = pd.DataFrame([], columns = event_tabs, index = emails)

for event in tqdm(event_tabs):
    df = db[event].dropna().drop_duplicates()

    attendee_event_clicks_matrix.loc[
        df.email, 
        event] = df.groupby(['email'])['total_clicks'].sum()
    
display(attendee_event_clicks_matrix.sample(20))
display(attendee_event_clicks_matrix.astype(float).describe())
```

## Ahora que representamos la información de esta manera, podemos enfocarnos en la asistencia. 

## En vez de quedarnos con el número total de clics, vamos a quedarnos únicamente con la incidencia: ¿hubo clic o no?

## Una vez que tenemos la información representada así, vamos a calcular el coeficiente de correlación de cada evento, con base a su asistencia.

## Si tengo un coeficiente alto (cercano a uno) indica que las personas que se conectaron a una sesión, probablemente también se conectaron a otra.

## Por otra parte, si el coeficiente es bajo (cercano a cero) quiere decir que ambos eventos tuvieron asistencias distintas. 

## el siguiente código va  darnos una tabla resumiendo exactamente esos coeficientes de correlación. 

```python
attendee_event_matrix = attendee_event_clicks_matrix > 0
corr_matrix = np.corrcoef(attendee_event_matrix.iloc[:, :].values.transpose())
corr_event_matrix = pd.DataFrame(corr_matrix, columns = event_tabs, index = event_tabs)

display(corr_event_matrix)
```

# Web Scrapping

## Como ejemplo, vamos a usar ésta página web: [Website to Scrape](http://quotes.toscrape.com/)

## Con el siguiente código, vamos a emular la navegación de esta página web, vamos a extraer cada cita y el nombre de cada autor, y los vamos a guardar en una tabla.

```python
import requests as rq
from bs4 import BeautifulSoup as bs
import warnings
warnings.filterwarnings("ignore")

url = 'http://quotes.toscrape.com/page/'
quotes = pd.DataFrame([], columns = ['Quote', 'Author'])

flag = True
i = 1
while flag:
    page = url + str(i)
    page = rq.get(page)
    soup = bs(page.content, "html.parser")
    
    if "No quotes found!" in str(soup):
        flag = False
        break
        
    for quote in soup.find_all('div', class_ = 'quote'):
        text = quote.find('span', class_ = "text")
        author = quote.find('small', class_ = 'author')
        row = pd.Series({'Quote':quote.text.strip(), 
                           'Author':author.text.strip()})
        
        quotes = quotes.append(row, ignore_index=True)
        
        print(f'{text.text} from {author.text}', '\n')
    
    i += 1
```

# Sesión 5: Tipos de Modelos de Inteligencia Artificial

## Algoritmo Genético

### Tenemos la siguiente función con 6 parámetros desconocidos:

$$ 44 = 4w_{1} + -2 w_{2} + 3.5w_{3} + 5w_{4} - 11w_{5} - 4.7w_{6} $$

### No tenemos más ecuaciones para despejar todas las variables, pero sabemos que la expresión tiene que ser igual a 44. 

### Con el siguiente código, vamos a correr un algoritmo genético simulando:

* ### 5 generaciones 
    - #### cada una con 5 individuos (posibles soluciones)
    - #### las cruzas entre individuos se harán una sola vez

```python
import pygad
import numpy

function_inputs = [4,-2,3.5,5,-11,-4.7] # Function inputs.
desired_output = 44 # Function output.

def fitness_func(ga_instance, solution, solution_idx):
    output = numpy.sum(solution*function_inputs)
    fitness = 1.0 / (numpy.abs(output - desired_output) + 0.000001)
    return fitness

num_generations = 5 # Number of generations.
num_parents_mating = 1

sol_per_pop = 5 # Number of solutions in the population.
num_genes = len(function_inputs)

last_fitness = 0
def on_generation(ga_instance):
    global last_fitness
    print(f"Generation = {ga_instance.generations_completed}")
    print(f"Fitness    = {ga_instance.best_solution(pop_fitness=ga_instance.last_generation_fitness)[1]}")
    print(f"Change     = {ga_instance.best_solution(pop_fitness=ga_instance.last_generation_fitness)[1] - last_fitness}")
    last_fitness = ga_instance.best_solution(pop_fitness=ga_instance.last_generation_fitness)[1]

ga_instance = pygad.GA(num_generations=num_generations,
                       num_parents_mating=num_parents_mating,
                       sol_per_pop=sol_per_pop,
                       num_genes=num_genes,
                       fitness_func=fitness_func,
                       on_generation=on_generation)

# Running the GA to optimize the parameters of the function.
ga_instance.run()

ga_instance.plot_fitness()

solution, solution_fitness, solution_idx = ga_instance.best_solution(ga_instance.last_generation_fitness)
print(f"Parameters of the best solution : {solution}")
print(f"Fitness value of the best solution = {solution_fitness}")
print(f"Index of the best solution : {solution_idx}")

prediction = numpy.sum(numpy.array(function_inputs)*solution)
print(f"Predicted output based on the best solution : {prediction}")

if ga_instance.best_solution_generation != -1:
    print(f"Best fitness value reached after {ga_instance.best_solution_generation} generations.")

# Saving the GA instance.
filename = 'genetic' # The filename to which the instance is saved. The name is without extension.
ga_instance.save(filename=filename)

# Loading the saved GA instance.
loaded_ga_instance = pygad.load(filename=filename)
loaded_ga_instance.plot_fitness()
```

## El siguiente código carga la imágen de estudio y crea un modelo de K-Means Clustering para automáticamente segmentar los colores en grupos de mayor incidencia. 

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import cv2
from collections import Counter
from skimage.color import rgb2lab, deltaE_cie76
import os

sab_RGBs=[
    [85,186,69],
    [255,168,37],
    [185,61,58],
    [165,59,57],
    [204,59,55],
    [219,220,221],
    [168,119,61],
    [222,178,129],
    [103,191,86],
    [221,183,101],
    [197,203,100],
    [135,158,80],
    [219,178,99],
    [85,71,78],
]

def pix2SqImg(rgb = [0, 0, 0]):
    clr = []
    for _ in range(200):
        clr_prov = []
        for _ in range(200):
            clr_prov.append(rgb)
            clr_prov = np.array(clr_prov)
            clr.append(clr_prov)
            clr = np.array(clr)
            plt.imshow(clr)
            
def RGB2HEX(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))

def get_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image

def get_colors(image, number_of_colors, show_chart):
    modified_image = cv2.resize(image, (150, 300), interpolation = cv2.INTER_AREA)
    modified_image = modified_image.reshape(modified_image.shape[0]*modified_image.shape[1], 3)
    clf = KMeans(n_clusters = number_of_colors)
    labels = clf.fit_predict(modified_image)
    counts = Counter(labels)
# sort to ensure correct color percentage
    counts = dict(sorted(counts.items()))
    center_colors = clf.cluster_centers_
# We get ordered colors by iterating through the keys
    ordered_colors = [center_colors[i] for i in counts.keys()]
    hex_colors = [RGB2HEX(ordered_colors[i]) for i in counts.keys()]
    rgb_colors = [ordered_colors[i] for i in counts.keys()]
    if (show_chart):
        plt.figure(figsize = (8, 6));
        _, _, text = plt.pie(counts.values(),
        labels = rgb_colors, #labels = hex_colors,
        colors = hex_colors,
        autopct='%.2f');
    return rgb_colors, text

plt.figure(figsize = (10, 20))
c, t = get_colors(get_image('sabRec.jpg'),len(sab_RGBs),True)
plt.tight_layout();
plt.savefig('colorPropSab.jpg');
plt.show();
```
